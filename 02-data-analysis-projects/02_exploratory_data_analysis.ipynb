## Apa itu Exploratory Data Analysis (EDA)?

Exploratory Data Analysis (EDA) adalah proses eksplorasi data untuk:
- Mengetahui karakteristik data
- Menemukan pola tersembunyi
- Mengidentifikasi outlier dan potensi masalah
- Membentuk hipotesis awal

Prinsip utama EDA:
> **“Let the data speak.”**

## Dataset yang Digunakan

Pada bab ini, kita menggunakan dataset sederhana data mahasiswa dengan atribut:
- Nama
- Nilai
- Jurusan
- Status Kelulusan

Dataset ini diasumsikan sudah melalui proses data cleaning.

### 1. Melakukan Import Library

```python
# Import Library
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns


# Konfigurasi tampilan
plt.style.use('default')
sns.set()
```

### 2. Load Dataset

```python
# Membuat DataFrame contoh

data = {
"Nama": ["Achmad", "Kamil", "Nurul", "Khassanah", "Fildza", "Mahreen", "Fillea"],
"Nilai": [85, 70, 78, 90, 60, 88, 75],
"Jurusan": ["Informatika", "Informatika", "Sistem Informasi", "Informatika", "Informatika", "Sistem Informasi", "Informatika"],
"Status Kelulusan": ["Lulus", "Tidak Lulus", "Lulus", "Lulus", "Tidak Lulus", "Lulus", "Lulus"]
}

df = pd.DataFrame(data)
df
```

### 3. Melihat Struktur data

Pada tahap awal EDA, kita perlu memahami:
- Jumlah baris dan kolom
- Tipe data setiap kolom
- Informasi missing values

```python
# Check informasi dataset
df.info()
```

### 4. Statistik Deskriptif

Statistik deskriptif membantu kita memahami:
- Nilai minimum dan maksimum
- Rata-rata (mean)
- Distribusi data

```python
# Check statistik numerik
df.describe()
```

### 5. Analisis Distribusi Cell

Distribusi data penting untuk:
- Mengetahui sebaran nilai
- Mendeteksi skewness dan outlier

```python
# Menampilkan Histogram Nilai Mahasiswa
plt.figure()
plt.hist(df['Nilai'])
plt.title('Distribusi Nilai Mahasiswa')
plt.xlabel('Nilai')
plt.ylabel('Frekuensi')
plt.show()
```

### 6. Analisis Per-Kategori

Analisis ini bertujuan untuk melihat perbedaan nilai berdasarkan `Jurusan`

```python
# Boxplot nilai per Jurusan
plt.figure()
sns.boxplot(x='Jurusan', y='Nilai', data=df)
plt.title('Distribusi Nilai per Jurusan')
plt.show()
```

### 7. Analisis Kelulusan

Mengetahui proporsi Mahasiswa yang `Lulus` dan `Tidak Lulus` akan memberikan gambaran performa akademik.

```python
# Countplot Status Kelulusan
plt.figure()
sns.countplot(x='Status Kelulusan', data=df)
plt.title('Jumlah Mahasiswa Berdasarkan Status Kelulusan')
plt.show()
```

### Insight Awal dari Data

Beberapa insight yang diambil:
- Mayoritas Mahasiswa memiliki Nilai diatas 75
- Jurusan Informatika memiliki variasi nilai lebih besar
- Mahasiswa dengan nilai dibawah 75 berpotensi memerlukan perhatian akademik

Insight tersebut belum final, tetapi menjadi dasar untuk analisis lanjutan atau dasar dalam pengambilan keputusan.
