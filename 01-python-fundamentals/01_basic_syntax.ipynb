## Variabel dalam Python

Variabel digunakan untuk menyimpan data di dalam memori komputer.

Python akan menentukan tipe data secara otomatis berdasarkan nilai yang diberikan ke variabel tersebut.

```python
# ---------------------------------------
# Contoh Variabel dengan Berbagai Tipe Data
# ---------------------------------------

# Variabel string (teks)
nama = "Kamil"

# Variabel integer (bilangan bulat)
umur = 30

# Variabel float (bilangan desimal)
tinggi_badan = 167.5

# Variabel boolean (True / False)
is_student = True
```

Penjelasan tipe data di atas:
- **string (str)** -> digunakan untuk teks
- **integer (int)** -> bilangan bulat
- **float** -> bilangan desimal
- **boolean (bool)** -> nilai logika (True atau False)

```python
# ---------------------------------------
# Menampilkan isi variabel
# ---------------------------------------

print("Nama             :", nama)
print("Umur             :", umur)
print("Tinggi Badan     :", tinggi_badan)
print("Status Mahasiswa :", is_student)
```

## Mengecek Tipe Data dengan type()

Dalam Data Science, sangat penting untuk mengetahui tipe data
karena setiap tipe data diproses dengan cara yang berbeda.

```python
# ---------------------------------------
# Mengecek tipe data setiap variabel
# ---------------------------------------

print(type(nama))
print(type(umur))
print(type(tinggi_badan))
print(type(is_student))
```

## String Formatting dengan f-string

f-string digunakan untuk menggabungkan teks dan variabel
secara rapih dan mudah dibaca.

```python
print(f"Nama saya {nama}")
print(f"Umur saya {umur} tahun")
print(f"Tinggi badan saya {tinggi_badan} cm")
print(f"Apakah masih mahasiswa? {is_student}")
```

## Input dari Pengguna

Fungsi `input()` digunakan untuk menerima data dari pengguna.
Perlu diperhatikan bahwa hasil dari `input()` selalu bertipe string.

```python
nama_input = input("Masukkan nama Anda: ")
umur_input = input("Masukkan umur Anda: ")

print(f"Nama: {nama_input}")
print(f"Umur: {umur_input}")

print(type(nama_input))
print(type(umur_input))
```

## Konvensi Penamaan Variabel

```python
# Contoh yang baik
total_belanja = 250000

# Contoh yang kurang baik
tb = 250000
```
