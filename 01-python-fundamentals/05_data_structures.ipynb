## Apa itu Data Structure?

Data structure adalah cara untuk menyimpan, mengorganisasi, dan mengelola data
agar dapat digunakan secara efisien.

Python menyediakan beberapa data structure bawaan yang disebut **collections**:
- List
- Tuple
- Set
- Dictionary

## 1. List

List adalah struktur data untuk menyimpan banyak nilai dalam satu variabel.

Karakteristik List:
- Berurutan (ordered)
- Bisa diubah (mutable)
- Bisa menyimpan berbagai tipe data

```python
# Membuat list
buah = ["apel", "pisang", "jeruk"]
angka = [10, 20, 30]
campuran = ["Achmad Kamil", 30, True, 170.5]

print(buah)
print(campuran)
```

### Mengakses dan Mengubah Data List

```python
# Mengakses data
print(buah[0])
print(buah[-1])

# Mengubah data
buah[1] = "mangga"
print(buah)
```

### Operasi Umum pada List

```python
buah.append("anggur")
buah.remove("apel")

print(buah)
print("Jumlah data:", len(buah))
```

### Looping pada List

```python
for item in buah:
    print(item)
```

## 2. Tuple

Tuple mirip dengan List, tetapi **tidak bisa diubah** (immutable).
Biasanya digunakan untuk data yang bersifat tetap.

```python
koordinat = (10, 20)
hari = ("Senin", "Selasa", "Rabu")

print(koordinat)
print(hari)
```

## 3. Set

Set adalah struktur data untuk menyimpan data **unik**.
Tidak berurutan dan tidak mengizinkan duplikasi.

```python
angka_unik = {1, 2, 3, 3, 4, 5}
print(angka_unik)
```

### Operasi Himpunan pada Set

```python
A = {1, 2, 3}
B = {3, 4, 5}

print("Union:", A | B)
print("Intersection:", A & B)
print("Difference:", A - B)
```

## 4. Dictionary

Dictionary menyimpan data dalam format **key : value**.
Struktur ini sangat umum digunakan dalam Data Science dan Data Engineering.

```python
mahasiswa = {
    "nama": "Achmad Kamil",
    "umur": 30,
    "tinggi": 170.5,
    "is_student": True
}

print(mahasiswa)
```

### Mengakses dan Mengubah Data Dictionary

```python
print(mahasiswa["nama"])
print(mahasiswa.get("umur"))

mahasiswa["umur"] = 22
mahasiswa["jurusan"] = "Teknik Informatika"

print(mahasiswa)
```

### Looping Dictionary

```python
for key, value in mahasiswa.items():
    print(key, ":", value)
```
